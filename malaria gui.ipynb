```python
from tkinter import*
from glob import glob
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tkinter import messagebox
from tkinter.filedialog import askopenfilename
from PIL import Image,ImageTk
import pyttsx3 as py


root=Tk()
root.title("malaria infected detection")
root.geometry("830x730")


f=Frame(root,relief=RIDGE,bd=6,bg='light slate gray')
f.place(x=0,y=0,width=830,height=130)
l=Label(f,text='Malaria Infected Detection',font=("ariel",40,'bold'),fg='white',bg='light slate gray')
l.place(x=110,y=25)

#main frame in center
frame=Frame(root,relief=GROOVE,bd=5,bg="light slate gray")
frame.place(x=0,y=131,width=830,height=470)

# image frame
frame1=Frame(frame,relief=GROOVE,bd=2)
frame1.place(x=200,y=4,width=430,height=330)

#button frame

frame2=Frame(frame,relief=GROOVE,bd=3,bg='light slate gray')
frame2.pack(side=BOTTOM)


#below frame
frame3=Frame(root,relief=GROOVE,bd=5,bg='light slate gray')
frame3.place(x=0,y=600,width=830,height=125)

image=[]
def upload():
    image.clear()
    file=askopenfilename()
    load=Image.open(file)
    res=load.resize((430,330))
    img=ImageTk.PhotoImage(res)
    label5.config(image=img)
    label5.image=img
    image.append(file)
    
def clear():
    label5.config(image=None)
    label5.image=None
    la.config(text=" ")
    
    
def predict():
    try:
        a=image[0]
        model1=load_model("maleria-model-43.h5")
        img=load_img(a,target_size=(224,224))
        test_img=img_to_array(img)
        test_img=test_img/255
        test_img=np.expand_dims(test_img,0)
        pred=model1.predict(test_img)
        pred=np.argmax(pred,1)
        if pred[0]==0:
            engine=py.init()
            engine.setProperty('rate',145)
            engine.say(f"Patient is infected by. Maleria")
            engine.runAndWait()
            la.config(text=f"Predicted output of the image is: Parasite",font=("ariel",25,'bold'),fg="white",bg="light slate gray")
        else:
            engine=py.init()
            engine.setProperty('rate',145)
            engine.say(f"Patient is not infected by. malaria")
            engine.runAndWait()
            la.config(text=f"Predicted output of the image is: Uninfected",font=("ariel",25,'bold'),fg="white",bg="light slate gray")
    except:
        messagebox.showerror("error","There is something wrong!")
        
def exit():
    if messagebox.askyesno("ask","do you want to quit?"):
        root.destroy()
    
la=Label(frame3,bg="light slate gray")
la.place(x=85,y=20)
    
label5=Label(frame1)
label5.place(x=0,y=0)
    
btn0=Button(frame,text="Upload Image Here",font=("ariel",14),command=upload)
btn0.place(x=300,y=340)

btn1=Button(frame2,text="Predict",font=("ariel",20),bg="green",fg="white",width=10,command=predict)
btn1.pack(side=LEFT)

btn2=Button(frame2,text="Clear",font=("ariel",20),fg="black",bg="yellow",width=10,command=clear)
btn2.pack(side=LEFT,padx=10)

btn3=Button(frame2,text="Exit",font=("ariel",20),fg="white",bg="red",width=10,command=exit)
btn3.pack(side=LEFT)

root.mainloop()
```
